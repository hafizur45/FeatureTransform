```python
def normalize_dataset(dataset):
    num_columns = len(dataset[0:0])

    for col_index in range(num_columns):
        # Extract values from the column
        column_values = [row[col_index] for row in dataset]

        # Calculate min and max values for the column
        min_val = min(column_values)
        max_val = max(column_values)

        # Normalize each value in the column
        for row in dataset:
            row[col_index] = (row[col_index] - min_val) / (max_val - min_val)

    return dataset

normalized_dataset = normalize_dataset(ar)
```
